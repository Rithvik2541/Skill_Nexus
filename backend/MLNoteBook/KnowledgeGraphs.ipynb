```python
def process_user(user_data):
    username = user_data.get('username')
    role = user_data.get('role')

    # Create or merge User node
    query = """
    MERGE (u:User {username: $username})
    SET u.role = $role
    """
    params = {'username': username, 'role': role}
    graph.query(query, params)

    # Process LeetCode Info
    leetcode_info = user_data.get('leetcodeInfo', [])
    if leetcode_info:
        contest_details = leetcode_info[0].get('userContestDetails', {})
        rating = contest_details.get('rating')
        attended_contests = contest_details.get('attendedContestsCount')
        query = """
        MATCH (u:User {username: $username})
        SET u.leetcode_rating = $rating,
            u.leetcode_contests = $attended_contests
        """
        params = {
            'username': username,
            'rating': rating,
            'attended_contests': attended_contests
        }
        graph.query(query, params)

    # Process GitHub Repositories
    github_info = user_data.get('githubInfo', [])
    if github_info:
        repositories = github_info[0].get('repositories', [])
        for repo_data in repositories:
            repo_name = repo_data.get('name')
            repo_description = repo_data.get('description', '')
            repo_language = repo_data.get('language', '')

            # Create Repository node and relationship
            query = """
            MATCH (u:User {username: $username})
            MERGE (r:Repository {name: $repo_name})
            SET r.description = $repo_description
            MERGE (u)-[:CREATED]->(r)
            """
            params = {
                'username': username,
                'repo_name': repo_name,
                'repo_description': repo_description
            }
            graph.query(query, params)

            # Process language as a Skill
            if repo_language:
                query = """
                MERGE (s:Skill {name: $skill_name})
                """
                params = {'skill_name': repo_language}
                graph.query(query, params)

                # Create relationships
                query = """
                MATCH (u:User {username: $username})
                MATCH (r:Repository {name: $repo_name})
                MATCH (s:Skill {name: $skill_name})
                MERGE (r)-[:USES]->(s)
                MERGE (u)-[:HAS_SKILL]->(s)
                """
                params = {
                    'username': username,
                    'repo_name': repo_name,
                    'skill_name': repo_language
                }
                graph.query(query, params)

    # Process LinkedIn Info
    linkedin_info = user_data.get('linkedinInfo', [])
    if linkedin_info:
        linkedin_profile = linkedin_info[0]
        # Process headline
        headline = linkedin_profile.get('headline', '')
        query = """
        MATCH (u:User {username: $username})
        SET u.linkedin_headline = $headline
        """
        params = {'username': username, 'headline': headline}
        graph.query(query, params)

        # Process positions
        positions = linkedin_profile.get('position', [])
        if positions is not None:
          for position_data in positions:
              company_name = position_data.get('companyName', '')
              title = position_data.get('title', '')
              start_year = position_data.get('start', {}).get('year')
              end_year = position_data.get('end', {}).get('year')

              # Create Position node and relationship
              query = """
              MATCH (u:User {username: $username})
              MERGE (p:Position {title: $title, company: $company_name})
              SET p.start_year = $start_year, p.end_year = $end_year
              MERGE (u)-[:WORKED_AT]->(p)
              """
              params = {
                  'username': username,
                  'title': title,
                  'company_name': company_name,
                  'start_year': start_year,
                  'end_year': end_year
              }
              graph.query(query, params)

        # Process skills
        skills = linkedin_profile.get('skills', [])
        if skills is not None:
          for skill_data in skills:
              skill_name = skill_data.get('name')
              query = """
              MERGE (s:Skill {name: $skill_name})
              """
              params = {'skill_name': skill_name}
              graph.query(query, params)

              # Create relationship
              query = """
              MATCH (u:User {username: $username})
              MATCH (s:Skill {name: $skill_name})
              MERGE (u)-[:HAS_SKILL]->(s)
              """
              params = {'username': username, 'skill_name': skill_name}
              graph.query(query, params)
```

```python
def process_user(user_data):
    username = user_data.get('username')
    role = user_data.get('role')

    # Create or merge User node
    query = """
    MERGE (u:User {username: $username})
    SET u.role = $role
    """
    params = {'username': username, 'role': role}
    graph.query(query, params)

    # Process LeetCode Info
    leetcode_info = user_data.get('leetcodeInfo', [])
    if leetcode_info:
        contest_details = leetcode_info[0].get('userContestDetails', {})
        rating = contest_details.get('rating')
        attended_contests = contest_details.get('attendedContestsCount')
        query = """
        MATCH (u:User {username: $username})
        SET u.leetcode_rating = $rating,
            u.leetcode_contests = $attended_contests
        """
        params = {
            'username': username,
            'rating': rating,
            'attended_contests': attended_contests
        }
        graph.query(query, params)

    # Process GitHub Repositories
    github_info = user_data.get('githubInfo', [])
    if github_info:
        repositories = github_info[0].get('repositories', [])
        for repo_data in repositories:
            repo_name = repo_data.get('name')
            repo_description = repo_data.get('description', '')
            repo_language = repo_data.get('language', '')

            # Create Repository node and relationship
            query = """
            MATCH (u:User {username: $username})
            MERGE (r:Repository {name: $repo_name})
            SET r.description = $repo_description
            MERGE (u)-[:CREATED]->(r)
            """
            params = {
                'username': username,
                'repo_name': repo_name,
                'repo_description': repo_description
            }
            graph.query(query, params)

            # Process language as a Skill
            if repo_language:
                query = """
                MERGE (s:Skill {name: $skill_name})
                """
                params = {'skill_name': repo_language}
                graph.query(query, params)

                # Create relationships
                query = """
                MATCH (u:User {username: $username})
                MATCH (r:Repository {name: $repo_name})
                MATCH (s:Skill {name: $skill_name})
                MERGE (r)-[:USES]->(s)
                MERGE (u)-[:HAS_SKILL]->(s)
                """
                params = {
                    'username': username,
                    'repo_name': repo_name,
                    'skill_name': repo_language
                }
                graph.query(query, params)

    # Process LinkedIn Info
    linkedin_info = user_data.get('linkedinInfo', [])
    if linkedin_info:
        linkedin_profile = linkedin_info[0]
        # Process headline
        headline = linkedin_profile.get('headline', '')
        query = """
        MATCH (u:User {username: $username})
        SET u.linkedin_headline = $headline
        """
        params = {'username': username, 'headline': headline}
        graph.query(query, params)

        # Process positions
        positions = linkedin_profile.get('position', [])
        if positions:  # Check if positions is not empty
            for position_data in positions:
                company_name = position_data.get('companyName', '')
                title = position_data.get('title', '')
                start_year = position_data.get('start', {}).get('year')
                end_year = position_data.get('end', {}).get('year')

                # Create Position node and relationship
                query = """
                MATCH (u:User {username: $username})
                MERGE (p:Position {title: $title, company: $company_name})
                SET p.start_year = $start_year, p.end_year = $end_year
                MERGE (u)-[:WORKED_AT]->(p)
                """
                params = {
                    'username': username,
                    'title': title,
                    'company_name': company_name,
                    'start_year': start_year,
                    'end_year': end_year
                }
                graph.query(query, params)

        # Process skills
        skills = linkedin_profile.get('skills', [])
        if skills:  # Check if skills is not empty
            for skill_data in skills:
                skill_name = skill_data.get('name')
                query = """
                MERGE (s:Skill {name: $skill_name})
                """
                params = {'skill_name': skill_name}
                graph.query(query, params)

                # Create relationship
                query = """
                MATCH (u:User {username: $username})
                MATCH (s:Skill {name: $skill_name})
                MERGE (u)-[:HAS_SKILL]->(s)
                """
                params = {'username': username, 'skill_name': skill_name}
                graph.query(query, params)
```

```python
def print_all_relationships():
    """Prints all relationships in the graph."""
    query = """
    MATCH (a)-[r]->(b)
    RETURN labels(a) AS from_labels, properties(a) AS from_properties,
           type(r) AS relationship,
           labels(b) AS to_labels, properties(b) AS to_properties
    """
    results = graph.query(query)
    print("All Relationships in the Graph:")
    for record in results:
        from_labels = record['from_labels']
        from_properties = record['from_properties']
        relationship = record['relationship']
        to_labels = record['to_labels']
        to_properties = record['to_properties']
        print(f"({from_labels}: {from_properties}) -[{relationship}]-> ({to_labels}: {to_properties})")
```

```python
def find_user_connections(username1, username2):
    # Query to find the shortest path between two users
    query = """
    MATCH path = shortestPath(
        (u1:User {username: $username1})-[*..4]-(u2:User {username: $username2})
    )
    RETURN path
    """
    params = {'username1': username1, 'username2': username2}
    results = graph.query(query, params)

    if not results:
        print(f"No connection found between {username1} and {username2}.")
        return

    print(f"Connections between {username1} and {username2}:")

    for record in results:
        path = record['path']
        print("Type of path:", type(path))
        print("Contents of path:", path)
```

```python
def find_user_connections(username1, username2):
    query = """
    MATCH path = shortestPath(
        (u1:User {username: $username1})-[*..4]-(u2:User {username: $username2})
    )
    RETURN [n IN nodes(path) | properties(n)] AS nodes,
           [r IN relationships(path) | {type: type(r), properties: properties(r)}] AS relationships
    """
    params = {'username1': username1, 'username2': username2}
    results = graph.query(query, params)

    if not results:
        print(f"No connection found between {username1} and {username2}.")
        return

    print(f"Connections between {username1} and {username2}:")

    for record in results:
        nodes = record['nodes']
        relationships = record['relationships']
        for i in range(len(relationships)):
            start_node = nodes[i]
            rel = relationships[i]
            end_node = nodes[i+1]
            start_label = start_node.get('username', start_node.get('name', ''))
            end_label = end_node.get('username', end_node.get('name', ''))
            rel_type = rel['type']
            print(f"({start_label}) -[{rel_type}]-> ({end_label})")
```

```python
def find_shared_skills(username1, username2):
    # Query to find shared skills between two users
    query = """
    MATCH (u1:User {username: $username1})-[:HAS_SKILL]->(s:Skill)<-[:HAS_SKILL]-(u2:User {username: $username2})
    RETURN s.name AS shared_skill
    """
    params = {'username1': username1, 'username2': username2}
    results = graph.query(query, params)

    shared_skills = [record['shared_skill'] for record in results]

    if not shared_skills:
        print(f"No shared skills between {username1} and {username2}.")
    else:
        print(f"Shared skills between {username1} and {username2}: {shared_skills}")
```

```python
def find_potential_mentors_with_connections(junior_username):
    query = """
    MATCH (junior:User {username: $junior_username})-[:WANTS_TO_LEARN]->(skill:Skill)
    MATCH (senior:User)-[:HAS_SKILL]->(skill)
    WHERE senior.role = 'Senior'
    WITH junior, senior, collect(DISTINCT skill.name) as matched_skills, count(DISTINCT skill) as skill_match_count
    OPTIONAL MATCH path = shortestPath((junior)-[*..4]-(senior))
    RETURN senior.username as mentor_username,
           matched_skills,
           skill_match_count,
           length(path) as path_length,
           [n IN nodes(path) | coalesce(n.username, n.name)] as path_nodes,
           [r IN relationships(path) | type(r)] as path_relationships
    ORDER BY skill_match_count DESC
    LIMIT 5
    """
    params = {'junior_username': junior_username}
    results = graph.query(query, params)

    mentors = []
    for record in results:
        mentor = {
            'username': record['mentor_username'],
            'matched_skills': record['matched_skills'],
            'skill_match_count': record['skill_match_count'],
            'path_length': record['path_length'],
            'path_nodes': record['path_nodes'],
            'path_relationships': record['path_relationships']
        }
        mentors.append(mentor)

    for mentor in mentors:
        print(f"Mentor Username: {mentor['username']}")
        print(f"Matched Skills: {mentor['matched_skills']}")
        print(f"Skill Match Count: {mentor['skill_match_count']}")
        print(f"Connection Path Length: {mentor['path_length']}")
        print("Connection Path:")
        for i in range(len(mentor['path_relationships'])):
            print(f"({mentor['path_nodes'][i]}) -[{mentor['path_relationships'][i]}]-> ({mentor['path_nodes'][i+1]})")
        print("\n")
```

```python
def compare_leetcode(username1, username2):
    query = """
    MATCH (u1:User {username: $username1}), (u2:User {username: $username2})
    RETURN u1.leetcode_rating AS rating1, u1.leetcode_contests AS contests1,
           u2.leetcode_rating AS rating2, u2.leetcode_contests AS contests2
    """
    params = {'username1': username1, 'username2': username2}
    result = graph.query(query, params)

    if not result:
        print(f"No Leetcode data found for users {username1} and {username2}.")
        return

    record = result[0]
    rating1 = record['rating1']
    contests1 = record['contests1']
    rating2 = record['rating2']
    contests2 = record['contests2']

    similarities = {}

    if rating1 and rating2:
        rating_diff = abs(rating1 - rating2)
        if rating_diff < 100:  # You can adjust the threshold
            similarities['rating'] = f"Both users have similar Leetcode ratings (difference of {rating_diff})."
    else:
        similarities['rating'] = "One or both users do not have Leetcode ratings."

    if contests1 and contests2:
        contests_diff = abs(contests1 - contests2)
        if contests_diff == 0:
            similarities['contests'] = f"Both users have attended the same number of contests ({contests1})."
        else:
            similarities['contests'] = f"Users have attended a similar number of contests (difference of {contests_diff})."
    else:
        similarities['contests'] = "One or both users do not have Leetcode contest data."

    print(f"Leetcode Similarities between {username1} and {username2}:")
    for key, value in similarities.items():
        print(f"- {value}")
```

```python
def compare_linkedin_skills(username1, username2):
    query = """
    MATCH (u1:User {username: $username1})-[:HAS_SKILL]->(s:Skill)<-[:HAS_SKILL]-(u2:User {username: $username2})
    RETURN collect(s.name) AS shared_skills
    """
    params = {'username1': username1, 'username2': username2}
    result = graph.query(query, params)

    shared_skills = result[0]['shared_skills'] if result else []

    if shared_skills:
        print(f"LinkedIn Skills shared between {username1} and {username2}: {shared_skills}")
    else:
        print(f"No shared LinkedIn skills between {username1} and {username2}.")
```

```python
def compare_linkedin_positions(username1, username2):
    query = """
    MATCH (u1:User {username: $username1})-[:WORKED_AT]->(p:Position)<-[:WORKED_AT]-(u2:User {username: $username2})
    RETURN p.title AS position_title, p.company AS company_name
    """
    params = {'username1': username1, 'username2': username2}
    results = graph.query(query, params)

    if results:
        print(f"LinkedIn Positions shared between {username1} and {username2}:")
        for record in results:
            print(f"- Title: {record['position_title']}, Company: {record['company_name']}")
    else:
        print(f"No shared LinkedIn positions or companies between {username1} and {username2}.")
```

```python
def compare_github_languages(username1, username2):
    query = """
    MATCH (u1:User {username: $username1})-[:HAS_SKILL]->(s:Skill)<-[:HAS_SKILL]-(u2:User {username: $username2})
    WHERE EXISTS {
        MATCH (u1)-[:CREATED]->(:Repository)-[:USES]->(s)
    } AND EXISTS {
        MATCH (u2)-[:CREATED]->(:Repository)-[:USES]->(s)
    }
    RETURN collect(DISTINCT s.name) AS shared_languages
    """
    params = {'username1': username1, 'username2': username2}
    result = graph.query(query, params)

    shared_languages = result[0]['shared_languages'] if result else []

    if shared_languages:
        print(f"GitHub Programming Languages shared between {username1} and {username2}: {shared_languages}")
    else:
        print(f"No shared GitHub programming languages between {username1} and {username2}.")
```

```python
def compare_github_repositories(username1, username2):
    query = """
    MATCH (u1:User {username: $username1})-[:CREATED]->(r:Repository)<-[:CREATED]-(u2:User {username: $username2})
    RETURN collect(r.name) AS shared_repositories
    """
    params = {'username1': username1, 'username2': username2}
    result = graph.query(query, params)

    shared_repositories = result[0]['shared_repositories'] if result else []

    if shared_repositories:
        print(f"GitHub Repositories shared between {username1} and {username2}: {shared_repositories}")
    else:
        print(f"No shared GitHub repositories between {username1} and {username2}.")
```

```python
def compare_users(username1, username2):
    print(f"\nComparing {username1} and {username2}:\n")

    # Compare Leetcode Data
    compare_leetcode(username1, username2)

    # Compare LinkedIn Skills
    compare_linkedin_skills(username1, username2)

    # Compare LinkedIn Positions
    compare_linkedin_positions(username1, username2)

    # Compare GitHub Programming Languages
    compare_github_languages(username1, username2)

    # Compare GitHub Repositories
    compare_github_repositories(username1, username2)
```
